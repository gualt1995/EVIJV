```python
from nltk.corpus import brown
news = brown.words(categories='news')
wordTagPairs = bigrams(news)
FreqDist([p for p in wordTagPairs])
```

```python
lntest = len([w for w in news if w == 'test'])
lncut = len([w for w in news if w == 'cut'])
lnski = len([w for w in news if w == 'ski'])
lntest
```
